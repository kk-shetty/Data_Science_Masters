#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a python library used for Data Analyisis and Vizualization. This is used to plot large set of data into a visual chart to analyze the sentiment of data quickly.
There are multiple graphs that we can plot using Matplotlibs Pyplot module.
1. **Line Plot** (`plt.plot()`): A line plot is used to represent data points connected by straight lines. It is commonly used to display trends or patterns over a continuous variable.
2. **Scatter Plot** (`plt.scatter()`): A scatter plot displays individual data points as dots on a graph. It is useful for visualizing the relationship between two continuous variables and identifying any patterns or correlations.
3. **Bar Plot** (`plt.bar()`): A bar plot (also known as a bar chart) represents data using rectangular bars of varying heights. It is often used to compare and display categorical data.
4. **Histogram** (`plt.hist()`): A histogram is used to represent the distribution of a dataset. It divides the data into bins and displays the frequency or count of values falling within each bin.
5. **Pie Chart** (`plt.pie()`): A pie chart represents data as slices of a circular pie. It is commonly used to show the proportion or percentage of different categories within a dataset.

#### Q2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated dataplot a scatter plot.

A **scatter plot** displays individual data points as dots on a graph. It is useful for visualizing the relationship between two continuous variables and identifying any patterns or correlations.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(x, y)
plt.xlabel("x-Value")
plt.ylabel("y-Value")
plt.title("Random Scatter Plot")
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
 * For line 1: `x = np.array([0, 1, 2, 3, 4, 5])` and `y = np.array([0, 100, 200, 300, 400, 500])`
 * For line 2: `x = np.array([0, 1, 2, 3, 4, 5])` and `y = np.array([50, 20, 40, 20, 60, 70])`
 * For line 3: `x = np.array([0, 1, 2, 3, 4, 5])` and `y = np.array([10, 20, 30, 40, 50, 60])`
 * For line 4: `x = np.array([0, 1, 2, 3, 4, 5])` and `y = np.array([200, 350, 250, 550, 450, 150])`

##### The `subplot()` function in Matplotlib is used to create a grid of subplots within a single figure. It allows us to create multiple plots (or subplots) and arrange them in rows and columns. This function is particularly useful when we want to visualize multiple plots side by side or compare different aspects of our data.

```python
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(4,1,1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(4, 1, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(4, 1, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(4, 1, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout spacing
plt.tight_layout()

# Display the subplots
plt.show()
```

#### Q4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A **bar plot** (also known as a bar chart) represents data using rectangular bars of varying heights. It is often used to compare and display categorical data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(4,4))
plt.bar(x=company, height=profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company v/s Profit - Bar Chart")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.barh(y=company, width=profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company v/s Profit - Horizontal Bar Chart")
plt.show()
```

#### Q5. What is a box plot? Why is it used? Using the following data plot a box plot.

* A **box plot**, also known as a **box-and-whisker plot**, is a graphical representation of a dataset's summary statistics and distribution. It displays key statistical measures such as the median, quartiles, and potential outliers in a concise and easy-to-understand manner.
* Box plots are used to visualize the distribution of numerical data and identify any significant features or differences between groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

data = [box1, box2]
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot')
plt.show()
```
