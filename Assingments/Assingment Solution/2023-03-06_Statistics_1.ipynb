#### Q5. Explain the concept of levels of measurement and give an example of a variable for each level.
**Level of Measurement** also called as **Scale of measurement** tells us how precisely variables are records. The level at which we measure a variable determines how we can analyze your data.
There are four levels of measurement:

1. **Nominal Scale data**: The data can only be categorized without any ordering. Example - city of birth, gender of students
2. **Ordinal Scale data**: We can categorizeank these variable in an order but we cannot say anything about the interval between the rankings. Example - Language ability, User reviews.
3. **Interval Scale data**: We can categorize, rank, and infer equal intervals between neighboring data points, but there is no true zero point. Example: Temperature values, IQ level
4. **Ratio Scale data**: We can categorize, rank, infer equal intervals betwee neghbouring data paoints and there is a true zero point. The difference and ratio are measurable. Example : height, weight an age.

#### Q9. Describe the difference between descriptive and inferential statistics. Give an example of each type of statistics and explain how they are used.
|Category|Descriptive Statistics|Inferential Statistics|
|---------|-----------------------|-----------------------|
|Purpose|Describe and summarize data|Make inferences and draw conclusions about a population based on sample data|
|Analysis|Analyzes and interprets the characteristics of a dataset|Uses sample data to make generalizations or predictions about a larger population|

#### Q10. What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.
**Measure of Centra; tendency** help us find the middle or the average of a dataset. The three most common measure of central tendency are the mode, median, and mean.
1. **mode**: the most frequently occurring value in the dataset. It’s possible to have no mode, one mode, or more than one mode. To find the mode, sort the dataset numerically or categorically and select the response that occurs most frequently.
2. **median**:  the value that’s exactly in the middle when it is ordered from low to high. To find the median, we first order all values from low to high. Then, find the value in the middle of the ordered dataset
3. **mean**: the sum of all values divided by the total number of values. It’s the most commonly used measure of central tendency because all values are used in the calculation. Outliers can significantly increase or decrease the mean when they are included in the calculation. Since all values are used to calculate the mean, it can be affected by extreme outliers. An outlier is a value that differs significantly from the others in a dataset.
