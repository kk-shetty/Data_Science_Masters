```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#### Scatter Plot
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
# Scatter plot
plt.scatter(x,y)
```

```python
# Beautification
plt.scatter(x,y)

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")
```

```python
# Beautification - color
plt.scatter(x,y, c='magenta') # Can also be HTML Color code

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")
```

```python
# Beautification - intensity
plt.scatter(x,y, c='red', alpha=.4)

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")
```

```python
# Beautification - Marker (Shape)
plt.scatter(x,y, c='red', alpha=.4, marker='^')

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")
```

```python
# Beautification - Marker (Shape)
plt.scatter(x,y, c='red', alpha=.4, marker='^')

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")

# Printing grid
plt.grid()
```

```python
#Controlling size of figure
plt.figure(figsize=(6,4))

# Beautification - Marker (Shape)
plt.scatter(x,y, c='red', alpha=.4, marker='^')

# axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Title
plt.title("Random Scatter Plot")

plt.grid()
```

```python
#### Line Graph
```

```python
# Generate a number between 1 to 10

x = np.linspace(1, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
#Controlling size of figure
plt.figure(figsize=(8,2))

# Plot by default will plot line graph
plt.plot(x,y, c='red', marker='.') #

# axis labels
plt.xlabel("x")
plt.ylabel("sin(x)")

# Title
plt.title("sin() wave")

plt.grid()
```

```python
### Bar graph
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f']
y = np.random.randint(20, 100, 6)
```

```python
plt.figure(figsize=(4, 3))
plt.bar(x,y)
plt.xlabel('categoty')
plt.ylabel('Volume')
plt.title("Sales by Category")
plt.grid(axis='y')
```

```python
# Bar horizontal
plt.figure(figsize=(4, 3))
plt.barh(x,y)
plt.xlabel('categoty')
plt.ylabel('Volume')
plt.title("Sales by Category")
plt.grid(axis='x')
```

```python
x = [3,4,5,6,7,8]
y = [10, 20, 30, 40, 50, 60]
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x,y, c='purple', marker='*')
plt.grid()
```

```python
# Freaquency - Histogram
```

```python
x = np.random.randint(0, 100, 500)
```

```python
plt.figure(figsize=(6, 4))
plt.hist(x, color='green', alpha=.4)
plt.grid()
```

```python
#### 3D Graph
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
