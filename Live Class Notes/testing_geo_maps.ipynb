```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the CSV file
data = pd.read_csv('dumps/cj_ej_po_prices.csv')
# Group the data by date
grouped_data = data.groupby('Day')
```

```python

# Initialize lists to store the data for each province
cj_min_prices = []
cj_max_prices = []
ej_min_prices = []
ej_max_prices = []

# Iterate over each group and extract the minimum and maximum prices for each province
for date, group in grouped_data:
    # print(f"{list(group[group['Province'] == 'Jawa Tengah']['Min Price'])}\n")
    if len(list(group[group['Province'] == 'Jawa Tengah']['Min Price'])) == 0:
        cj_min_prices.append(0)
    else:
        cj_min_prices.append(list(group[group['Province'] == 'Jawa Tengah']['Min Price'])[0])

    if len(list(group[group['Province'] == 'Jawa Tengah']['Max Price'])) == 0:
        cj_max_prices.append(0)
    else:
        cj_max_prices.append(list(group[group['Province'] == 'Jawa Tengah']['Max Price'])[0])

    if len(list(group[group['Province'] == 'Jawa Timur']['Min Price'])) == 0:
        ej_min_prices.append(0)
    else:
        ej_min_prices.append(list(group[group['Province'] == 'Jawa Timur']['Min Price'])[0])

    if len(list(group[group['Province'] == 'Jawa Timur']['Max Price'])) == 0:
        ej_max_prices.append(0)
    else:
        ej_max_prices.append(list(group[group['Province'] == 'Jawa Timur']['Max Price'])[0])
```

```python
# Plot the range bar chart
plt.figure(figsize=(12, 6))

plt.bar(grouped_data.groups.keys(), cj_max_prices, label='CJ', alpha=0.5)
plt.bar(grouped_data.groups.keys(), cj_min_prices, label='CJ', alpha=0.5)

plt.bar(grouped_data.groups.keys(), ej_max_prices, label='EJ', alpha=0.5)
plt.bar(grouped_data.groups.keys(), ej_min_prices, label='EJ', alpha=0.5)
```
