```python
import random
import json

# List of Indian cities
indian_cities = ['Mumbai', 'Delhi', 'Bangalore', 'Kolkata', 'Chennai', 'Hyderabad', 'Ahmedabad', 'Pune', 'Jaipur', 'Lucknow']

# List of Indian names
indian_names = ['Aarav', 'Aisha', 'Akash', 'Ananya', 'Arjun', 'Avni', 'Dev', 'Esha', 'Gaurav', 'Isha', 'Karthik', 'Krishna', 'Maya', 'Neha', 'Rahul', 'Riya', 'Rohan', 'Sahana', 'Sanjay', 'Tanvi']

# Generate 50 JSON records
records = []
for _ in range(50):
    name = random.choice(indian_names)
    age = random.randint(21, 60)
    city = random.choice(indian_cities)
    salary = random.randint(15000, 150000)

    record = {'name': name, 'age': age, 'City': city, 'Salary': salary}
    records.append(record)

# Convert records to JSON
json_records = json.dumps(records, indent=4)

# Print the JSON records
print(json_records)
```
