```python
cust_df = pd.DataFrame(customers)
prod_df = pd.DataFrame(products)
order_df = pd.DataFrame(orders)
```

```python
pd.merge(order_df, cust_df, how = 'left', left_on = 'cust id', right_on = 'id')
```

```python
# Select only required columns
pd.merge(order_df, cust_df, how = 'left', left_on = 'cust id', right_on = 'id')[['id_x', 'date', 'cust id', 'name', 'prod id']]
```

```python
# rename selected columns
pd.merge(order_df, cust_df, how = 'left', left_on = 'cust id', right_on = 'id')[['id_x', 'date', 'cust id', 'name', 'prod id']].rename(columns={'id_x':'order id'})
```
