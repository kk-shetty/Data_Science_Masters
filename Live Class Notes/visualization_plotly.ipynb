```python
import plotly.graph_objects as go
import numpy as np
import seaborn as sns
```

```python
# Creating a canvas
fig = go.Figure()
arr1 = np.random.randint(1,100, 1000)
arr2 = np.random.randint(200, 700,1000)
fig.add_trace(go.Scatter(x=arr1, y=arr2, mode='markers'))
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=tips.day, y=tips.total_bill.value_counts()))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=tips.tip))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode='markers', marker_size = tips['size']*5 ))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill, y=tips.tip, mode='markers', z = tips['size'] ))
```
