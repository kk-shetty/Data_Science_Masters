#### SMOTE (Synthetic Minority Oversampling Technique)

```python
from sklearn.datasets import make_classification # make_clkassification is used to produce dataset 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a data set of features and target - return a tuple (independant features, target)
x, y = make_classification(
    n_samples=1000, 
    n_redundant=0,
    n_features=2, 
    n_clusters_per_class=1,
    weights=[0.9],
    random_state=90)
```

```python
df_independent = pd.DataFrame(x, columns=['f1', 'f2'])
df_target = pd.DataFrame(y, columns=['target'])
df_final = pd.concat([df_independent, df_target], axis = 1)
df_final.head()
```

```python
# Check how imabalance the data is
df_final['target'].value_counts()
```

```python
# Plot a scatter plot to look how data is scattered
plt.figure(figsize=(10, 10))
plt.scatter(x=df_final['f1'], y=df_final['f2'], c=df_final['target'])
plt.show()
```

```python
# imblearn needs to be installed
from imblearn.over_sampling import SMOTE
```

```python
# Transfer the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(df_final[['f1', 'f2']], df_final['target'])
```

```python
X.shape
```

```python
oversampled_df = pd.concat([X, y], axis=1)
oversampled_df.head()
```

```python
oversampled_df['target'].value_counts()
```
