#### Data Interpolation
* Process of estimating unknown values within a dataset based on known values.

##### 1. Linear Interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
## Interpolate the dxata using linear interpolation
x_new = np.linspace(1,5,10) # Create new X values
y_interep = np.interp(x_new, x, y)
```

```python
plt.scatter(x_new,y_interep)
```

##### 2. Cubic Interpolation with Scipy

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
```

```python
## Create a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
## Interpolate the data
x_new = np.linspace(1,5,18)
y_interp = f(x_new)
```

```python
plt.scatter(x_new, y_interp)
```

##### 3. Polynomial Interpolation

```python
# create sample data
l1 = [1,5,10,12,17]
l2 = [x**2 - (3 * x) + 4 for x in l1]

x = np.array(l1)
y = np.array(l2)

plt.scatter(x,y)
```

```python
## Interopolate the data using polynomial interpolation
p=np.polyfit(x,y,2)

x_new = np.linspace(1,17, 20)
y_interp = np.polyval(p, x_new)

plt.scatter(x_new, y_interp)
```
