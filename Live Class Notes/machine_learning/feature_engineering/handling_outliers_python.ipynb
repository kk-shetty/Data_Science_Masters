```python
## Minimum, Maximum, Median, Q1 and Q3
```

```python
import numpy as np
```

```python
marks = np.random.randint(28,100, 32)
min, Q1, median, Q3, max = np.quantile(marks,[0.0, 0.25, 0.5, 0.75, 1.0])
print(f"1. Minimum - {min}\n2. Q1 - {Q1}\n3. Median - {median}\n4. Q3 - {Q3}\n5. Maximum - {max}")
```

```python
import seaborn as sns

sns.boxplot(marks)
```
