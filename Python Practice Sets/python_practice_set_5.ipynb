```python
''' Q1.
Can you create a Python program that reads a text file and counts the number of words containedwithin it? 
The program should be designed to read the file and break it down into individual words, using spaces, punctuation marks, and other delimiters to separate the words.
It should then count the number of words found in the file and display this count as output. 
The program should be flexible enough to work
with different text files and should be able to handle a variety of formatting and punctuation styles.
'''
import string

def get_words(file_name):
    words = []
    with open(file_name, 'r') as file:
        for line in file:
            line = line.translate(str.maketrans('', '', string.punctuation))
            words.extend(line.split())
    return words
```

```python
words = len(get_words('sample.txt'))
print(f"Word Count = {words}")
```

```python
''' Q2. 
Could you help me create a Python program that can find and display the longest word in a text file?
The program should be able to read any text file and separate its contents into individual words, taking into account various delimiters like spaces, punctuation marks, and other characters.
Then, it should compare the length of each word and determine which one is the longest, and finally, print that word as output.
The program should be versatile enough to work with various text files and be able to handle
different formatting styles and punctuation.
'''
def get_longest_word(file_name):
    max_word = ''
    max_word_len = 0
    for word in get_words(file_name):
        if len(word) > max_word_len:
            max_word_len = len(word)
            max_word = word
    return (max_word, max_word_len)
```

```python
file_name = 'sample.txt'
max_word, max_len = get_longest_word(file_name)
print(f"Longest Word : {max_word}\nLength : {max_len} characters")
```

```python
''' Q3. 
Write a Python program to read a text file and print out the most frequent word(s) in the file.
'''

def word_count(file_name):
    words = get_words(file_name)
    word_freaquency = {}
    for word in words:
        if word.lower() in word_freaquency:
            word_freaquency[word.lower()] += 1
        else:
            word_freaquency[word.lower()] = 1
    return word_freaquency
```

```python
file_name = 'sample.txt'
max_accurence = 0
words = []

for key, value in word_count(file_name).items():
    if value > max_accurence:
        max_accurence = value
        words = [key]
    elif value == max_accurence:
        words.append(key)
print(words)
print(max_accurence)
```

```python
''' Q4.
How can you use Python to count the number of rows in a CSV file?
'''

import csv

file_name = 'sample.csv'
with open(file_name, 'r') as file:
    rows = csv.reader(file)
    num_of_rows = sum(1 for row in rows)
    print(f'Number of Rows: {num_of_rows}')
    
```

```python
''' Q5.
How can you use Python to calculate the average of a specific column in a CSV file?
'''
file_name = 'sample.csv'
with open(file_name, 'r') as file:
    rows = csv.reader(file)
    total = 0
    count = 0

    for index, row in enumerate(rows):
        # Initialize the columns index or get based on header
        if index == 0:
            col_index = row.index('c')
        else:
            try:
                total += float(row[col_index])
                count += 1
            except Exception as e:
                print(e)
    average = total/count
    print(average)
```

```python
''' Q8.
Can you provide a Python code to read a binary file and display the data in hexadecimal format.
'''

# Generate a binary file:
with open("sample.binary", 'wb') as file:
    # Generate a sequence of bytes
    byte_sequence = bytes(range(256))

    # Repeat the sequence multiple times
    num_repetitions = 100
    file.write(byte_sequence * num_repetitions)

# Read and display binary
file_name = 'sample.binary'
with open(file_name, "rb") as file:
    binary_data = file.read()
    print([format(byte, '02X') for byte in binary_data])
```

```python
''' Q9.
Can you write a Python code to read a Comma-Separated Values (CSV) file, apply a specific condition to each row, and create a new CSV file that contains only the rows that satisfy the condition?
For example, if the CSV file contains information about products and their prices, you may want to create a new CSV file that only includes the products that are within a certain price range. 
The program should be able to read the CSV file, compare the values in each row to the specified condition, and write the rows that meet the criteria to a new CSV file.
'''

input_file = 'product_price.csv'
filtered_file = 'high_price_products.csv'

with open(input_file, 'r') as file:
    data = csv.reader(file)
    filtered_rows = []
    for index, row in enumerate(data):
        # This is required if the CSV has a header and to fetch index of columns 
        if index == 0:
            ind_of_price = row.index('price')
            filtered_rows.append(row) # If we need to include header row in filtered csv
        else:
            try:
                price = float(row[ind_of_price])
                if price >= 20:
                    filtered_rows.append(row)
            except Exception as e:
                print(e)
print(filtered_rows) 

with open(filtered_file, 'w') as file:
    csv.writer(file).writerows(filtered_rows)
```
